```python
import pandas as pd
dataset = pd.read_csv('./EdStatsData.csv');
```

```python
# country name
# country code
# indicator name
# indicator code

# years
```

```python
indicators       = dataset.iloc[:, 2].values
```

```python
dictionaire = {}
for i in indicators:
    words = i.split(' ')
    for w in words:
        if (not w in dictionaire):
            dictionaire[w] = 0
        dictionaire[w] += 1
        
```

```python
forbiddenWords = ['of', 'in', '(%)', 'by', '(%).', 'the', 'with', 'EGRA:']
```

```python
bests = []
for (k, v) in dictionaire.items():
    # Si la clé (donc le mot) est un mot interdit, je saute l'ajout
    if k in forbiddenWords:
        continue
        
    index = 0
    while (index < len(bests)):
        if (bests[index][1] < v):
            break;
        index += 1
    bests.insert(index, (k, v))
        
```

```python
for i in range(10):
    print(bests[i])
```

```python
import matplotlib.pyplot as plt
```

```python
plottedX = []
plottedY = []
for i in range(0, 10):
    plottedX.append(bests[i][0])
    plottedY.append(bests[i][1])
    
plt.xlabel('Words')
plt.ylabel('Number')

plt.bar(plottedX, plottedY)
plt.xticks(rotation=-90)
plt.title('Histogram of top 10 used words')
plt.show()
```

```python
years       = dataset.iloc[:, 4:(4 + 2050 - 1970 + 1 + 7)].values
```

```python
xValues = []
yValues = []
for yearNum in range(2017 - 1970 + 1 + 7): # on ajoute les 7 valeurs supplémentaires qui votn de 5 en 5
    if (yearNum + 1970 > 2017): # après 2017 on compte de 5 en 5
        xValues.append(str((yearNum - (2017 - 1970 + 1)) * 5 + 2020)) 
        # on rapène year num à zéro avec (yearNum - (2017 - 1970 + 1)
        # on multiplie par 5
        # on ajoute 2020 car on commence à 2020
    else:
        xValues.append(str(yearNum + 1970))
    n = 0
    for v in range(len(years)):
        if (years[v][yearNum] == years[v][yearNum]): # not nan
            n += 1
    yValues.append(n)
        
```
